```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Streams.csv")
data.head()
```

```python
data.info()
```

```python
unique_song=data[data.Streams>2]['Song'].unique()
unique_song.sort()
unique_song
```

```python
song_stream_case = []
for i in unique_song :
    song_stream_case.append(data[data.Streams>2][data['Song']==i].Streams.sum())
```

```python
print (set(zip(unique_song,song_stream_case)))
```

```python
plt.barh(unique_song,song_stream_case)
plt.show()
```

```python
plt.figure(figsize=(10,20))
plt.pie(song_stream_case)
plt.legend(unique_song, loc='best')
plt.show()
```
